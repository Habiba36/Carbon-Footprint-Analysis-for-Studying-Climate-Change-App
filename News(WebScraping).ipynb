```python
from flask import Flask,flash, request,redirect, url_for,render_template, jsonify
import requests
from bs4 import BeautifulSoup
import json
```

```python
app = Flask(__name__)
```

```python
@app.route('/News')
def Scraping():
    listOfLists = []
    # WorldWide News
    URL_WW = "https://www.nbcnews.com/climate-in-crisis"
    page_WW = requests.get(URL_WW,headers={'User-Agent':'Mozilla/5.0'}).text
    soup_WW = BeautifulSoup(page_WW, 'lxml')
    titles_WW = soup_WW.find_all('h2',class_='wide-tease-item__headline')
    titles_list_WW = []
    d={}
    for title in titles_WW:
      titles_list_WW.append(title.text.strip())
    descriptions_WW = soup_WW.find_all('div',class_='wide-tease-item__description')
    descriptions_list_WW = []
    for description in descriptions_WW:
      descriptions_list_WW.append(description.text.strip())
    links_WW = soup_WW.find_all('div', class_='wide-tease-item__info-wrapper flex-grow-1-m')
    links_list_WW = []
    for link in links_WW:
      a = link.find_all('a')[1]
      links_list_WW.append(a['href'])
    
    worldWidelist = []
    for i in range(len(descriptions_list_WW)):
        x ={'title':titles_list_WW[i],'description':descriptions_list_WW[i],'link':links_list_WW[i]}
        worldWidelist.append(x)
        
    # Egypt News
    URL_EG = "https://www.egypttoday.com/Tag/5726/Climate-Change"
    page_EG = requests.get(URL_EG,headers={'User-Agent':'Mozilla/5.0'}).text
    soup_EG = BeautifulSoup(page_EG, 'lxml')
    titles_EG = soup_EG.find_all('div',class_='BOrderBoxSection')
    titles_list_EG = []
    for title in titles_EG:
      h3 = title.find('h3')
      a = h3.find('a')
      titles_list_EG.append(a.text.strip())
    descriptions_EG = soup_EG.find_all('div',class_='BOrderBoxSection')
    descriptions_list_EG = []
    for description in descriptions_EG:
      p = description.find('p').text.strip()
      descriptions_list_EG.append(p)  
    links_EG = soup_EG.find_all('div',class_='BOrderBoxSection')
    links_list_EG = []
    for link in links_EG:
      h3 = link.find('h3')
      a = h3.find('a')
      links_list_EG.append("https://www.egypttoday.com"+a['href'])

    Egyptlist = []
    for i in range(len(descriptions_list_EG)):
        y ={'title':titles_list_EG[i],'description':descriptions_list_EG[i],'link':links_list_EG[i]}
        Egyptlist.append(y)

    listOfLists.append(worldWidelist)
    listOfLists.append(Egyptlist)
    jsonObj = json.dumps(listOfLists,ensure_ascii=False)
    print(jsonObj)
    
    return jsonObj;

if __name__ == "__main__":  
    app.run(debug = False,host='192.168.56.1', port=5000)
```
